```python
train_df = pd.read_csv("train_pre.csv",index_col=0)
test_df = pd.read_csv("test_pre.csv",index_col=0)
X_train = train_df.drop("Survived",axis=1)
y_train = train_df["Survived"]
X_test = test_df
```

### 简单模型

使用最简单的逻辑回归模型，我们在验证集上可以得到81的正确率

```python
#简单逻辑回归模型
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split,cross_validate,KFold,cross_val_score

# 按一定比例拆分
train,verfy = train_test_split(train_df,test_size=.2)

X_train_1 = train.drop("Survived",axis=1)
y_train_1 = train["Survived"]

X_verfy_1 = verfy.drop("Survived",axis=1)
y_verfy_1 = verfy["Survived"]


logreg = LogisticRegression()
logreg.fit(X_train_1, y_train_1)

logreg.score(X_verfy_1,y_verfy_1)
```

我们在验证集上可以得到80%左右的正确率

同时我们可以看看各个特征对结果的影响,可以看到性别与是否幸存的关联性最大

```python
log_coef = pd.DataFrame(X_train.columns)
log_coef.columns = ["Features"]
log_coef["coef"] = pd.Series(logreg.coef_[0])

sns.barplot(x="Features",y="coef",data=log_coef)
```

### 交叉验证

原始数据中已经有训练集和测试集，我们再进行模型训练时应该遵循一个原则：尽量不要碰测试数据，最大程度保证测试数据评估模型的准确性

所以我们先将训练集拆分成训练集和验证集，拆分有两种方式：
1. 按某个比例拆分
比如训练集占70，验证集占30%（上一步）
2. 交叉验证
a. 留一交叉验证（leave-one-out）：每次从个数为N的样本集中，取出一个样本作为验证集，剩下的N-1个作为训练集，重复进行N次。最后平均N个结果作为泛化误差估计。
b. k-fold交叉验证：将数据分为K份，每次取一份作为验证集，其余k-1份作为训练集，重复进行K次，最后平均k个结果作为泛化误差估计


```python
# K-fold
#可以生成索引
#kf = KFold(n_splits=5,shuffle=False)

from sklearn.model_selection import cross_val_score

log_clf = LogisticRegression(random_state=23)
cross_val_score(log_clf, X_train, y_train, cv=3, scoring="accuracy")
```

使用交叉验证对于提高准确率似乎没有很大的提升,但是相比当个的模型预测有更好的稳定性。之前没有交叉验证模型的准确率在0.78-0.81之间震荡


到目前为止，我们只是用 预测正确的数/总的样本数 这样的指标来衡量模型的好坏，在实际工作中，只是用这个指标还无法满足要求。这时我们引入一个新的概念，混淆矩阵。

### 混淆矩阵
#### (不到最后一步，不要碰你的测试数据)
可以先用交叉验证代替评估模型效果

混淆矩阵中每一行代表真实值，列代表预测值，其中T表示是判定对了还是判定错了，P和N表示预测值是什么







最好的预测结果是：TP 和 TN 有值，其他两个是0 

准确率：Accuracy= （TP+TN）/ （TP+TN+FP+FN）
精确率： Precision = TP / (TP + FP)
召回率： Recall = (TP) / ( TP + FN)
比如



举个例子：有100个样本，其中有10个高风险人
预测模型1：所有人都是高风险
则：召回率是100% 因为我把所有高风险人都挑出来了，但是付出的代价就是把90个低风险的人也判定为高风险，此时精确率是： 10% 
宁可错杀100不可放过1个

### f1分数

Precision体现了模型对负样本的区分能力，Precision越高，模型对负样本的区分能力越强；Recall体现了模型对正样本的识别能力，Recall越高，模型对正样本的识别能力越强。F1 score是两者的综合，F1 score越高，说明模型越稳健。

F1 = 2 * (precision * recall) / (precision + recall)


```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_predict

y_predict = cross_val_predict(log_clf,X_train,y_train,cv=3)
confusion_matrix(y_train,y_predict)
```

### 平衡精确率和召回率


对于二分类，我们可以调整阈值改变判定的结果,首先通过decision_function获取每个判别类的得分，而不仅是判别结果，再通过改变阈值得到新的判别结果

```python

y_predic_score = cross_val_predict(log_clf,X_train,y_train,cv=3,method="decision_function")

from sklearn.metrics import precision_recall_curve

#计算不同的threshold值时，precision和 recall如何变化
precisions,recalls,thresholds = precision_recall_curve(y_train,y_predic_score[:,1])

def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], "b--", label="Precision")
    plt.plot(thresholds, recalls[:-1], "g-", label="Recall")
    plt.xlabel("Threshold")
    plt.legend(loc="upper left")
    plt.ylim([0, 1])
plot_precision_recall_vs_threshold(precisions, recalls, thresholds)
plt.show()
```

### ROC曲线

ROC曲线是衡量二分类效果的另外一个工具，里面有2个概念
TPR: 即召回率  TP/(TP+FN)
FPR:  负样本被误判为正样本的比例  FP / ( FP + TN)

### 图形解释

1. 虚线表示完全随机时的ROC曲线
2. 阈值越高召回率越高
3. 一个好的分类器应该离随机越远越好（向左上角方向）
4. 根据3可以用一个指数来衡量分类器好坏，即曲线右下方的面积



```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_train, y_predic_score[:,1])

def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([0, 1, 0, 1])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    
plot_roc_curve(fpr, tpr)
plt.show()

from sklearn.metrics import roc_auc_score

roc_auc_score(y_train, y_predic_score[:,1])
#0.84918884947645368
```

### 网格搜索

当你不知道使用什么超参数时，可以试试sklearn提供的网格搜索，网格搜索会尝试你提供的所有参数，比较每一种参数下模型的准确率


逻辑回归的参数主要有两个方面，一个是正则化的选择。
还有是多分类的设置，多分类怎么转化成二分类，在转换过程中产生的参数，还有是设置权值的参数，



```python
from sklearn.model_selection import  GridSearchCV
params = {"penalty":["l1","l2"],"solver":[  "liblinear"]}

logistic = LogisticRegression()

grid_log = GridSearchCV(logistic,params,cv = 5)

grid_log.fit(X_train,y_train)

grid_log.best_params_
```

看起来网格搜索在逻辑回归种用处不大，但是对于决策树或者随机森林来说，网格搜索的好处就显而易见了
