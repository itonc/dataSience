### 地理编码
地理编码是指通过一个文本地址，找到其对应的经纬度,我们利用百度的地理编码接口完成这个事情
接口使用参考文档：http://lbs.baidu.com/index.php?title=webapi/guide/webservice-geocoding

注：这里也可以用pyecharts中的Geo().get_coordinate( "深圳市新洲花园") 方法

```python
def getLocation(address):
    lat = 0 
    lng = 0 
    url = "http://api.map.baidu.com/geocoding/v3/"
    parms = {
    'address':address,
    'output':'json',
    'ak':'PCiCHRK1snUjMmjOzTBEnBGx2Px8lhQ9',
    'callback':'showLocation'}
    
    queryString=parse.urlencode(parms)
    u = request.urlopen(url +'?'+queryString)

    try:
        resp=u.read().decode('utf-8').replace('showLocation&&showLocation(','').replace(')','')
        location = json.loads(resp).get("result").get("location")
        lat = location.get("lat")
        lng = location.get("lng")
    except Exception:
        pass
    
    return  [lng,lat]
    



location = getLocation('深圳市福田区华富街道福莲花园')
print(location)
```

### 简单地图撒点
本次我们的地图将要使用pyecharts包
pyechart参考文档：https://pyecharts.org/#/zh-cn/geography_charts?id=geo%ef%bc%9a%e5%9c%b0%e7%90%86%e5%9d%90%e6%a0%87%e7%b3%bb


我们先用pyechars 做一个简单的地图撒点，即把深圳市所有确诊患者出现过的地方在深圳地图上描绘出来


```python
# 先把经纬度处理成一个json串

loc_dic = {}
loc_tup = []
indexs = []
for area,loc in zip(df.address,df.loaction):
    indexs.append(area)
    loc_tup.append((area,loc))
    loc_dic[area] = loc
    

values = np.ones(121).tolist()

#深圳所有的区
district = df.area.unique().tolist()
dis_values = np.zeros(len(district)).tolist()

#存入一个本地文件
with open("loc_json.json",'w') as f:
    json.dump(loc_dic,f)
    
    
    
    
```

```python
from pyecharts.globals import ChartType
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.options import LabelOpts,ItemStyleOpts,LegendOpts


def geo_simple() -> Geo:
    c = (
        Geo()
        .add_schema(maptype="深圳",
                    label_opts=LabelOpts(is_show=True,font_size=16),
                    itemstyle_opts = ItemStyleOpts(color = 'rgb(242,242,232)'))
        .add_coordinate_json("loc_json.json")
        .add('确诊小区', [list(z) for z in zip(indexs,values)]  ,symbol_size = 8)
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(title_opts=opts.TitleOpts(title="深圳新型肺炎分布图",pos_left='center'),
                         legend_opts = opts.LegendOpts(pos_right = 100,pos_top = 60))
    )
    return c

geo_simple().render_notebook()
```

### 问题

1. 没有体现出时间维度作用

接下来，我们会在这个图上加上随时间变化各个地方确诊人的变化


```python
# 用publishdate作为字典的索引

pdates = df.publishdate.unique().tolist()

dates_dic = {}

for date in pdates:
    values = []
    dates_dic[date] = values
    dfs = df[df.publishdate == date]
    for area,loc in zip(dfs.address,dfs.loaction):
        values.append((area,loc))
    

```

```python
from pyecharts.charts import Timeline



def geo_effectscatter() -> Timeline:
    t1 = Timeline().add_schema(play_interval = 1200,
                              is_timeline_show = False,
                              is_auto_play = True)
    for i in sorted(dates_dic.keys()):
        c = (
            Geo()
            .add_schema(maptype="深圳",
                   label_opts=LabelOpts(is_show=True,font_size=16),
                   itemstyle_opts = ItemStyleOpts(color = 'rgb(242,242,232)'))
        
            .add('确诊小区',
                data_pair = dates_dic[i],
                symbol_size = 8)
            .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
            .set_global_opts(title_opts=opts.TitleOpts(title="深圳新型肺炎小区分布图",
                                                       pos_left='center',
                                                       subtitle="{}".format(i),
                                                       item_gap = 20,
                                                      subtitle_textstyle_opts=opts.TextStyleOpts(font_size=26,
                                                                                                color = 'blue')),
                         legend_opts = opts.LegendOpts(pos_right = 100,pos_top = 60))
        
            )
        t1.add(c,"{}".format(i))
    return t1
   
geo_effectscatter().render_notebook()
```


### 关系图

接下来我们获取深圳卫健委公布的病例间关系图可视化

```python
from pyecharts import options  as opts
relation = pd.read_csv('relation.csv',header=None)
relation.columns = ['start','end','relaType']
relation.tail()
maxNode = 178

nodes = []
for n in range(1,maxNode+1):
    nodes.append(opts.GraphNode(name=str(n),symbol_size=10)) 

links = []
for start,end,v in zip(relation.start,relation.end,relation.values):
    links.append(opts.GraphLink(source=str(start),target = str(end),value=v))
```

```python
def graph_with_edge_opts() -> Graph:
    nodes_data = nodes
    links_data = links
    c = (
        Graph()
        .add(
            "",
            nodes_data,
            links_data,
            repulsion=4000,
            edge_label=opts.LabelOpts(
                is_show=False,
                position="middle",
                
            ),
        )
        .set_global_opts(
            title_opts=opts.TitleOpts(title="Graph-GraphNode-GraphLink-WithEdgeLabel")
        )
    )
    return c
graph_with_edge_opts().render_notebook()
```
